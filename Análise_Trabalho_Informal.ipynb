## 3.3 Tratando os tipos de dados

### 3.3.1 Substituindo valores das colunas ( 'posicao_ocupacao', 'grupos_ocupacao' e 'atividade_ramo_negocio_agregado') por nomes

```python
# Substituir valores na coluna 'posicao_ocupacao'
dados_01['posicao_ocupacao'] = dados_01['posicao_ocupacao'].replace({2: 'conta própria'})

# Substituir valores na coluna 'grupos_ocupacao'
mapeamento_grupos_ocupacao = {
    1: 'técnicas e afins',
    2: 'administrativas',
    3: 'agropecuária',
    4: 'indústria de transformação',
    5: 'comércio',
    6: 'transporte e comunicação',
    7: 'prestação de serviços',
    8: 'outras',
    9: 'ignorado'
}
dados_01['grupos_ocupacao'] = dados_01['grupos_ocupacao'].replace(mapeamento_grupos_ocupacao)

# Substituir valores na coluna 'atividade_ramo_negocio_agregado'
mapeamento_ramo_negocio = {
    1: 'agrícola',
    2: 'indústria de transformação',
    3: 'indústria de construção',
    4: 'outras atividades industriais',
    5: 'comércio de mercadorias',
    6: 'prestação de serviços',
    7: 'serviços auxiliares da atividade...',
    8: 'transporte e comunicação',
    9: 'social',
    10: 'administração pública',
    11: 'outras atividades ou não declarar'
}
dados_01['atividade_ramo_negocio_agregado'] = dados_01['atividade_ramo_negocio_agregado'].replace(mapeamento_ramo_negocio)

# Substituir valores na coluna 'ultimo_grau_frequentado'
mapeamento_ultimo_grau = {
    1: 'elementar (primário)',
    2: 'médio primeiro ciclo (ginasial)',
    3: 'médio segundo ciclo (científico)',
    4: 'primeiro grau',
    5: 'segundo grau',
    6: 'superior',
    7: 'mestrado/doutorado',
    8: 'alfab de adultos',
    9: 'pré-escolar ou creche'
}
dados_01['ultimo_grau_frequentado'] = dados_01['ultimo_grau_frequentado'].replace(mapeamento_ultimo_grau)
dados_01
```

### 3.3.2 Transformando os valores do tipo folat64 para o formato int, sem casas decimais

```python
# Lista das colunas a serem convertidas
colunas_a_converter = ['ocupacao_semana', 'atividade_ramo_negocio_semana', 'renda_mensal_ocupacao_principal','renda_mensal_ocupacao_principal_deflacionado', 'renda_mensal_dinheiro','renda_mensal_dinheiro_deflacionado','horas_trabalhadas_semana','horas_trabalhadas_todos_trabalhos', 'grau_frequentado', 'anos_estudo']

# Use um loop para converter cada coluna para inteiros
for coluna in colunas_a_converter:
    dados_01[coluna] = dados_01[coluna].fillna(0).astype(int)

# Verifique o resultado
dados_01


```

### - Selecionando apenas as colunas desejadas

```python
# Usando a função loc para selecionar as colunas desejadas
grupos_ocupacao = dados_01.loc[:, ["ano", "sigla_uf", "posicao_ocupacao", "grupos_ocupacao","atividade_ramo_negocio_agregado","renda_mensal_dinheiro","ultimo_grau_frequentado"]]
grupos_ocupacao
```

### - Inserindo a coluna Variação do PIB que contém a variação do PIB brasileiro entre 1981 e 2001

```python
# Dicionário com as variações do PIB para cada ano
variacoes_pib = {
    1981: -4.3,
    1982: 0.8,
    1983: -2.9,
    1984: 5.4,
    1985: 7.8,
    1986: 7.5,
    1987: 3.5,
    1988: -0.1,
    1989: 3.2,
    1990: -4.3,
    1991: 1.0,
    1992: -0.5,
    1993: 4.9,
    1994: 5.8,
    1995: 4.2,
    1996: 2.2,
    1997: 3.4,
    1998: 0.3,
    1999: 0.5,
    2000: 4.4,
    2001: 1.4
}

# Usando o método loc para criar a nova coluna com base no ano
grupos_ocupacao['Variação do PIB'] = grupos_ocupacao['ano'].map(variacoes_pib)
grupos_ocupacao
```

### - Inserindo coluna Situação Econômica com base na coluna Variação do PIB

```python
# Crie uma função para definir a situação econômica com base na variação do PIB
def definir_situacao_economica(variacao):
    if variacao > 5.0:
        return "Alto Crescimento"
    elif 2.0 <= variacao <= 5.0:
        return "Crescimento Moderado"
    elif variacao < 2.0 and variacao >= 0.0:
        return "Baixo Crescimento"
    else:
        return "Retração Econômica"

# Use a função para criar a nova coluna "Situação Econômica"
grupos_ocupacao["Situação Econômica"] = grupos_ocupacao["Variação do PIB"].apply(definir_situacao_economica)
grupos_ocupacao
```

### - Quantidade de empregos por conta própria de acordo com a situação econômica

```python
# Conte os elementos da coluna 'posicao_ocupacao' por 'Situação Econômica'
contagem = grupos_ocupacao.groupby(['Situação Econômica', 'posicao_ocupacao']).size().unstack().fillna(0)

# Crie um gráfico de barras
contagem.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Paired')

# Defina os rótulos dos eixos
plt.xlabel('Situação Econômica')
plt.ylabel('Contagem')

# Defina o título do gráfico
plt.title('Contagem de posicao_ocupacao por Situação Econômica')

# Mostre o gráfico
plt.legend(title='posicao_ocupacao', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Suponhamos que você tenha o DataFrame grupos_ocupacao com as colunas posicao_ocupacao, grupos_ocupacao, renda_mensal_dinheiro, ultimo_grau_frequentado e ano.

# Definir o valor de corte em salários mínimos.
valor_corte_salario_minimo = 180

# Definir um dicionário de mapeamento para os níveis de instrução.
nivel_instrucao = {
    'elementar (primário)': 'baixo',
    'médio primeiro ciclo (ginasial)': 'baixo',
    'médio segundo ciclo (científico)': 'alto',
    'primeiro grau': 'baixo',
    'segundo grau': 'alto',
    'superior': 'alto',
    'mestrado/doutorado': 'alto',
    'alfab e adultos': 'baixo',
    'pré-escolar ou creche': 'baixo'
}

# Aplicar o mapeamento para criar uma nova coluna que categoriza o grau de instrução.
grupos_ocupacao['categoria_instrucao'] = grupos_ocupacao['ultimo_grau_frequentado'].map(nivel_instrucao)

# Criar um DataFrame para armazenar os resultados.
resultados = pd.DataFrame(columns=['Ano', 'Grupo de Ocupação', 'Contagem Baixo Rendimento', 'Contagem Melhor Rendimento'])

# Definir os grupos de ocupação.
grupos = ['Administrativas', 'Agropecuária', 'Comércio', 'Indústria de Transformação', 'Outras', 'Prestação de Serviços', 'Transporte e Comunicação', 'Técnicas e Afins']

# Calcular a contagem para cada grupo de ocupação e ano.
for ano in grupos_ocupacao['ano'].unique():
    for grupo in grupos:
        # Filtrar os empregos conta própria no grupo de ocupação e ano atual.
        empregos_grupo_ano = grupos_ocupacao[
            (grupos_ocupacao['grupos_ocupacao'] == grupo) &
            (grupos_ocupacao['ano'] == ano)
        ]

        # Filtrar os empregos no grupo de baixo rendimento e baixo grau de instrução.
        baixo_rendimento_baixo_grau = empregos_grupo_ano[
            (empregos_grupo_ano['posicao_ocupacao'] == 'Conta própria') &
            (empregos_grupo_ano['renda_mensal_dinheiro'] <= valor_corte_salario_minimo) &  # Até o valor de corte em salários mínimos.
            (empregos_grupo_ano['categoria_instrucao'] == 'baixo')
        ]

        # Filtrar os empregos no grupo de melhor rendimento e maior grau de instrução.
        melhor_rendimento_maior_grau = empregos_grupo_ano[
            (empregos_grupo_ano['posicao_ocupacao'] == 'Conta própria') &
            (empregos_grupo_ano['renda_mensal_dinheiro'] > valor_corte_salario_minimo) &  # Acima do valor de corte em salários mínimos.
            (empregos_grupo_ano['categoria_instrucao'] == 'alto')
        ]

        # Calcular as contagens.
        contagem_baixo_rendimento_baixo_grau = len(baixo_rendimento_baixo_grau)
        contagem_melhor_rendimento_maior_grau = len(melhor_rendimento_maior_grau)

        # Adicionar os resultados ao DataFrame de resultados.
        resultados = resultados.append({'Ano': ano, 'Grupo de Ocupação': grupo, 'Contagem Baixo Rendimento': contagem_baixo_rendimento_baixo_grau, 'Contagem Melhor Rendimento': contagem_melhor_rendimento_maior_grau}, ignore_index=True)

# Exibir os resultados.
resultados;
```
